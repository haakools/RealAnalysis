```python
import numpy as np
import matplotlib.pyplot as plt
```

### Plot of $f(x)=|x|$ and its fourier series
### $\frac{\pi}{2} -\frac{4}{\pi}\sum_{n=0}^\infty \frac{\cos \left[(2n+1)x\right]}{(2n+1)^2}$

```python
def fourier(x,N):
    # Fourier series for f(x) = |x| between [-pi, pi]
    a0 = np.pi/2
    series = np.array([(4/np.pi)*np.cos((2*n+1)*x)/(2*n+1) for n in range(N)])
    return a0-series.sum()
```

```python

x = np.linspace(-np.pi, np.pi, 1000)
N = 3

fig, ax = plt.subplots(figsize=(10,6))
plt.rcParams['text.usetex'] = True

ax.plot(x, np.abs(x), c='k', label="$f(x)=|x|$")
ax.set_ylabel("$f(x)$", size=20)
ax.set_xlabel("$x$", size=20)

ax.set_yticks([-np.pi, 0, np.pi])
ax.set_yticklabels(["$\pi$", "$0$", "$\pi$"], color="k", size=20)

for n in range(N):
    ax.plot(x,
            np.array([fourier(x[i],n) for i in range(len(x))]),
           label="Fourier Series n={}".format(n))
    ax.legend(loc='lower center', bbox_to_anchor=(0.5, -0.2),
          ncol=2, fancybox=True, shadow=True, fontsize=14)
    
```
